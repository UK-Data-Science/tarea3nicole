```python
# Contador para las apariciones de 'HDI rank'
count = 0

# Iterando a través de la lista
for i, col in enumerate(column_names):
    if col == 'HDI rank':
        count += 1
        # Si es la segunda aparición, renombrar
        if count == 2:
            column_names[i] = 'HDI rank.1'
            break

column_names
```

Removing unnecessary rows in each of the DFS

```python
#Creating a function to clean the rows

def clean_rows(df, column_name):
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
        cleaned_df = df.dropna(subset=[column_name])
        return cleaned_df
```

```python
# Cleaning each Data Frame using the function
allDFs = [clean_rows(df, 'HDI rank') for df in allDFs]
```

```python
# rename the columns
newCol={0:col[0],1:col[1],2:col[2],3:col[3],4:col[4],5:col[5],6:col[6],7:col[7],8:col[8],9:col[9]}
df.rename(columns=newCol,inplace=True)
df.info()
```

```python
# select the useful columns
keep1=['FECHA_FALLECIMIENTO']
dfSome1=df.loc[:,keep1].copy()

# transform the useful dates

dfSome1.FECHA_FALLECIMIENTO=pd.to_datetime(dfSome1.FECHA_FALLECIMIENTO).dt.to_period('M')
dfSome1
```

```python
# ddd a count column
dfSome1.loc[:,'NUM_FALLECIDOS']=1

# aggregate the column
varsAggregated=['NUM_FALLECIDOS']
functionForVars='sum'
dfSome1=dfSome1.groupby('FECHA_FALLECIMIENTO')[varsAggregated].agg(functionForVars)
dfSome1
```

```python
dfSome1.reset_index(drop=False, inplace=True)
```
